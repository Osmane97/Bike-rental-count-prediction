#### Avoid Data Leakage
##### To prevent information leakage between training and evaluation:
- **Training set:** All records from the year 2011  
- **Evaluation set (validation):** First three quarters of 2012  
- **Holdout set (test set):** Last quarter of 2012


```python
# Sort by date (important to avoid shuffling issues)
data_day =data_day.sort_values(by = 'dteday')
data_hour = data_hour.sort_values(by = 'dteday')
```

```python
# Define cutoff dates
cutoff_date = '2012-01-01'
holdout_day =  '2012-10-01'
```

```python
# Train: before 2012
train_data_day = data_day[data_day['dteday']< cutoff_date]
#valid: before Q4 /2012
valid_data_day = data_day[(data_day['dteday']>= cutoff_date) & (data_day['dteday']< holdout_day)]
# Test: after Q3 2012
holdout_data_day = data_day[data_day['dteday']>= holdout_day]
```

```python
# Train: before 2012
train_data_hour = data_hour[data_hour['dteday']< cutoff_date]
#valid: before Q4 /2012
valid_data_hour = data_hour[(data_hour['dteday']>= cutoff_date) & (data_hour['dteday']< holdout_day)]
#test: after Q3 /2012
holdout_data_hour = data_hour[data_hour['dteday']>= holdout_day]
```
